```python
import pandas as pd

import matplotlib.pyplot as plt

a=pd.read_csv("/content/methane.csv")
plt.figure(figsize=(14,10))
plt.title("Methane emission")
plt.plot(a['Species'],a['Year'],color="green",label="1967")
plt.plot(a['Species'],a['Year2'],color="red",label="2010")
plt.xlabel("Species --> ")
plt.ylabel("Year --> ")
plt.legend()
plt.show()
```

```python
import pandas as pd

import matplotlib.pyplot as plt

a=pd.read_csv("/content/animals-slaughtered-for-meat-india.csv")
plt.figure(figsize=(14,10))
plt.title("Animals Slaughtered for meat India")
plt.plot(a['Year'],a['cattle'],color="green",label="cattle")
plt.plot(a['Year'],a['goat'],color="yellow",label="goat")
plt.plot(a['Year'],a['chicken'],color="red",label="chicken")
plt.plot(a['Year'],a['pig'],color="blue",label="pig")
plt.plot(a['Year'],a['lamb'],color="cyan",label="lamb")
plt.plot(a['Year'],a['duck'],color="magenta",label="duck")
plt.xlabel("Years --> ")
plt.ylabel("animals slaughtered --> ")
plt.legend()
plt.show()
```
